Question 2 (40%): Using the Coffee stack exchange data and posts (both question and answers),
implement a Boolean retrieval system (Convert all the text to lowercase). and search the
following queries:<br/>
>a. espresso [information need: how to make espresso]<br/>
b. turkish + coffee (and) [information need: making Turkish coffee]<br/>
c. turkish + coffee (or) [information need: making Turkish coffee]<br/>
d. persian + coffee (not-and) [information need: making Persian coffee]


```python
# 2. read in the posts and create the Boolean Retrieval system

from post_parser_record import PostParserRecord
post_reader = PostParserRecord("Posts_Coffee.xml")

queryWordMatrixDict = {"espresso":[],"coffee":[],"turkish":[],"persian":[]}

# iterate through each question post, convert the post body to lower case 
# and then split the post body into separate words

for question_id in post_reader.map_questions:
  question = post_reader.map_questions[question_id]
  fullQuestion = question.title + " " + question.body
  questionLowercase = fullQuestion.lower()
  questionWords = questionLowercase.split(" ")

  # Now see if any of those question words are the query word

  for specificWord in questionWords:
    # if the word is one of the queryWords, add the ID to that array

    if specificWord == 'espresso':
      if not queryWordMatrixDict['espresso']:
        queryWordMatrixDict[specificWord].append(question_id)

      if queryWordMatrixDict['espresso'][-1] != question_id:
        queryWordMatrixDict[specificWord].append(question_id)


    if specificWord == 'coffee':
      if not queryWordMatrixDict['coffee']:
        queryWordMatrixDict[specificWord].append(question_id)

      if queryWordMatrixDict['coffee'][-1] != question_id:
        queryWordMatrixDict[specificWord].append(question_id)


    if specificWord == 'turkish' :
      if not queryWordMatrixDict['turkish']:
        queryWordMatrixDict[specificWord].append(question_id)

      if queryWordMatrixDict['turkish'][-1] != question_id:
        queryWordMatrixDict[specificWord].append(question_id)


    if specificWord == 'persian' :
      if not queryWordMatrixDict['persian']:
        queryWordMatrixDict[specificWord].append(question_id)

      if queryWordMatrixDict['persian'][-1] != question_id:
        queryWordMatrixDict[specificWord].append(question_id)


#print(queryWordMatrixDict)

# now do answers

for answer_id in post_reader.map_just_answers:
  #print(answer_id)
  answer = post_reader.map_just_answers[answer_id]
  answerLowercase = answer.body.lower()
  #print(answerLowercase)
  answerWords = answerLowercase.split(" ")
  #print(answerWords)

  # Now see if any of those answer words are the query word
  # add to the word 

  for eachWord in answerWords:

    if eachWord == 'espresso':

      if not queryWordMatrixDict['espresso']:
        queryWordMatrixDict[eachWord].append(answer_id)

      if queryWordMatrixDict['espresso'][-1] != answer_id:
        queryWordMatrixDict[eachWord].append(answer_id)


    if eachWord == 'coffee':
      if not queryWordMatrixDict['coffee']:
        queryWordMatrixDict[eachWord].append(answer_id)

      if queryWordMatrixDict['coffee'][-1] != answer_id:
        queryWordMatrixDict[eachWord].append(answer_id)


    if eachWord == 'turkish' :
      if not queryWordMatrixDict['turkish']:
        queryWordMatrixDict[eachWord].append(answer_id)

      if queryWordMatrixDict['turkish'][-1] != answer_id:
        queryWordMatrixDict[eachWord].append(answer_id)


    if eachWord == 'persian' :
      if not queryWordMatrixDict['persian']:
        queryWordMatrixDict[eachWord].append(answer_id)

      if queryWordMatrixDict['persian'][-1] != answer_id:
        queryWordMatrixDict[eachWord].append(answer_id)
 

# sort the results so that they are in order
queryWordMatrixDict['espresso'].sort()
queryWordMatrixDict['turkish'].sort()
queryWordMatrixDict['coffee'].sort()
queryWordMatrixDict['persian'].sort()



print(queryWordMatrixDict)

```
